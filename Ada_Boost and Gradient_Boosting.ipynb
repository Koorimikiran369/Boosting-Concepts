```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('bank_oversample_new.csv')
pd.set_option('display.max_columns',None)
df.sample(5)
```

```python
df.shape
```

```python
print(df['result'].describe())
plt.figure(figsize=(9, 8))
sns.distplot(df['result'], color='g', bins=100, hist_kws={'alpha': 0.4});
```

```python
df.describe().T
```

```python
X = df.drop('result',axis = 1)
y = df['result']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 40)
print(X_train.shape,X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier #clasification tree
```

## AdaBoost

```python
model_ada = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth = 10),n_estimators=50,learning_rate = 1)
```

```python
model_ada.fit(X_train,y_train)
```

```python
import innomatics
```

```python
y_pred = model_ada.predict(X_test)
y_prob = model_ada.predict_proba(X_test)
```

```python
innomatics.classification_metric(y_test,y_pred,y_prob[:,1],n=50,
                                label = ['no','yes'], verbose = True)
```

## GradientBoosting

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model_grad = GradientBoostingClassifier(n_estimators=500,max_depth=10)
```

```python
model_grad.fit(X_train,y_train)
```

```python
y_pred = model_grad.predict(X_test)
y_prob = model_grad.predict_proba(X_test)
```

```python
innomatics.classification_metric(y_test,y_pred,y_prob[:,1],n=50,
                                label = ['no','yes'], verbose = True)
```
